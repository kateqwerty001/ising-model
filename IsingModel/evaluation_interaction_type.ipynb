```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from Graphs.Graphs import LatticeGraph, ErdosRenyiGraph, CompleteGraph, BarabasiAlbertGraph, WattsStrogatzGraph
from IsingModel.IsingModel import IsingModel
import warnings
import pandas as pd
from scipy.ndimage import gaussian_filter1d
warnings.filterwarnings("ignore", category=RuntimeWarning)

np.random.seed(42)
seed = 0
```

# Complete graph

--------------------

```python
complete25 = CompleteGraph(25).get_ajacency_matrix()
complete49 = CompleteGraph(49).get_ajacency_matrix()
complete100 = CompleteGraph(100).get_ajacency_matrix()
```

#### ferromagnetic interaction + complete graph

```python
T_values = np.linspace(0, 100.0, 200)
records = []

for T in T_values:
    for graph, N in zip([complete25, complete49, complete100], [25, 49, 100]):
        magnetizations = []

        for _ in range(10):
            model = IsingModel(graph, "Complete", T=T, interaction_type="ferro",
                               steps=50_000, burn_in=10_000, sample_every=100,
                               seed=seed)
            _, mean_m, _ = model.simulate()
            magnetizations.append(mean_m)

        median_m = np.median(magnetizations)

        records.append({
            "N": N,
            "mean_m": median_m,
            "T": T,
            "interaction_type": "ferro"
        })

results = pd.DataFrame(records)
results.to_csv("exp_interaction_type/results_complete_ferro.csv", index=False)
```

#### antiferromagnetic interaction + complete graph

```python
T_values = np.linspace(0, 10.0, 100)
records = []

for T in T_values:
    for graph, N in zip([complete25, complete49, complete100], [25, 49, 100]):
        magnetizations = []

        for _ in range(3):
            model = IsingModel(graph, "Complete", T=T, interaction_type="antiferro",
                               steps=50_000, burn_in=10_000, sample_every=100,
                               seed=seed)
            _, mean_m, _ = model.simulate()
            magnetizations.append(mean_m)

        median_m = np.median(magnetizations)

        records.append({
            "N": N,
            "mean_m": median_m,
            "T": T,
            "interaction_type": "antiferro"
        })

results = pd.DataFrame(records)
results.to_csv("exp_interaction_type/results_complete_antiferro.csv", index=False)
```

#### spin_glass + complete graph

```python
T_values = np.linspace(0, 20, 100)
records = []

for T in T_values:
    for graph, N in zip([complete25, complete49, complete100], [25, 49, 100]):
        magnetizations = []

        for _ in range(3):
            model = IsingModel(graph, "Complete", T=T, interaction_type="spin_glass",
                               steps=50_000, burn_in=10_000, sample_every=100,
                               seed=seed)
            _, mean_m, _ = model.simulate()
            magnetizations.append(mean_m)

        median_m = np.median(magnetizations)

        records.append({
            "N": N,
            "mean_m": median_m,
            "T": T,
            "interaction_type": "spin_glass"
        })

results = pd.DataFrame(records)
results.to_csv("exp_interaction_type/results_complete_spin_glass.csv", index=False)
```

```python
ferro = pd.read_csv("exp_interaction_type/results_complete_ferro.csv")
antiferro = pd.read_csv("exp_interaction_type/results_complete_antiferro.csv")
spin_glass = pd.read_csv("exp_interaction_type/results_complete_spin_glass.csv")

colors = {25: "#C6BD17", 49: "#4C86A8", 100: "#8E3B46"}

fig, axes = plt.subplots(1, 3, figsize=(21, 6), dpi=200, sharey=False)

titles = [
    "Ferromagnetic interaction",
    "Antiferromagnetic interaction",
    "Spin glass interaction"
]

datasets = [ferro, antiferro, spin_glass]

for ax, data, title in zip(axes, datasets, titles):
    for N in [25, 49, 100]:
        subset = data[data["N"] == N].sort_values("T")
        T = subset["T"].values
        mean_m = subset["mean_m"].values

        smooth_m = gaussian_filter1d(mean_m, sigma=2)
        dM_dT = np.gradient(smooth_m, T)
        Tc_index = np.argmax(-dM_dT)
        Tc = T[Tc_index]

        ax.plot(T, mean_m, label=f"N = {N}", color=colors[N], linewidth=2, marker='.')
        ax.axvline(Tc, color=colors[N], linestyle='--', linewidth=1.5, alpha=0.7)
        ax.text(Tc + 0.05, 0.05, f"Tc ‚âà {Tc:.2f}", rotation=90, color=colors[N],
                fontsize=20, verticalalignment='bottom')

    ax.set_title(f"Complete Graphs:\n{title}", fontsize=20)
    ax.set_xlabel("Temperature T", fontsize=18)
    ax.set_ylim(0, 1)
    ax.grid(True)
    ax.tick_params(axis='both', which='major', labelsize=18)

axes[0].set_ylabel("Magnetization", fontsize=18)
axes[2].legend(title="Lattice size", loc="upper right", prop={'size': 16})

plt.tight_layout()
plt.show()
```
